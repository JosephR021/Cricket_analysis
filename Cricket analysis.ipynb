**Match Summary**

```python
#Assigning the filepath to variables

data_summary = r"E:\Own_Project\Cricket analysis\data-analytics-project-for-beginners\t20_json_files\t20_json_files\t20_wc_match_results.json"

# opening the json file

with open(data_summary) as f_summary:
    data_summary = json.load(f_summary)
    
#creating a dataframe to save the json file
    
df_summary = pd.DataFrame(data_summary[0]['matchSummary'])
```

```python
# renmaming the scorecard column to matchid(primary key)

df_summary.rename({'scorecard' : 'matchid'}, axis =1 ,inplace=True)
```

**Batting summary**

```python
#Assigning the filepath to variables

data_bat = r"E:\Own_Project\Cricket analysis\data-analytics-project-for-beginners\t20_json_files\t20_json_files\t20_wc_batting_summary.json"
 
#opening the json file(this will only show one match result)

with open(data_bat) as f_bat:
    data_bat = json.load(f_bat)
    
#TO make all records to display

all_records =[]

for rec in data_bat:
    all_records.extend(rec['battingSummary'])
    
#creating a dataframe to save the json file

df_bat = pd.DataFrame(all_records)
```

```python
df_bat.head()
```

```python
#adding a column out/notout in df_bat 

df_bat['out/notout'] = df_bat.dismissal.apply(lambda x: 'out' if len(x) >0 else 'notout')
```

```python
df_bat.head(20)
```

```python
#Cleaning the player names which contains some special characters

import re

#this is to replace all the special characters

pattern = re.compile(r'[^\x00-\x7F]+')

df_bat.batsmanName = df_bat.batsmanName.str.replace(pattern, '')

df_bat.dismissal = df_bat.dismissal.str.replace(pattern,'')
```

```python
df_bat.head(20)
```

```python
# creating dictionary to make connections between the tables

match_ids_dict = {}

#Storing the combination of team1 and team2 in all possibilities

for index, row in df_summary.iterrows():
    key1 = row.team1 + ' Vs ' + row.team2
    key2 = row.team2 + ' Vs ' + row.team1
    
    match_ids_dict[key1] = row.matchid
    match_ids_dict[key2] = row.matchid
```

```python
match_ids_dict
```

```python
#adding a new column (matchid) in df_bat table using the match_ids_dict

df_bat['matchid'] = df_bat.match.map(match_ids_dict)
```

```python
df_bat.head(11)
```

**Bowling Summary**

```python
df_bowl.tail()
```

```python
#Cleaning the unwanted characters in the bowlerName column

df_bowl.bowlerName = df_bowl.bowlerName.str.replace(pattern,'')
```

```python
#Adding a new column matchid by mapping the dictionary

df_bowl['matchid'] = df_bowl.match.map(match_ids_dict)
```
